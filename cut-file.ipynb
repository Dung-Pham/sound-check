### Import thư viện

```python
from pydub import AudioSegment
from pydub.silence import detect_nonsilent
from pathlib import Path
from pprint import pprint

```

### Hàm lấy tên file .mp3

```python
# 1. lấy danh sách tên các file trong 5 folder nhạc cụ

def get_name_mp3_of_type (type):
    folder = Path(f"./sound-source/{type}")
    return [f.name for f in folder.iterdir() if f.is_file() and f.suffix == ".mp3"]

def get_mp3 (types):
    for name in types :
        types[name].extend(get_name_mp3_of_type(name))
```

```python
def cut_mp3 (type, name, count):
    # 1. Load file âm thanh (hỗ trợ wav, mp3, flac…)
    audio = AudioSegment.from_file(f"./sound-source/{type}/{name}", format="mp3")

    # 2. Phát hiện các đoạn có âm thanh (nonsilent)
    #    - min_silence_len: tối thiểu độ dài khoảng im lặng (ms) để coi là silence
    #    - silence_thresh: ngưỡng (dBFS) dưới ngưỡng này sẽ coi là silence
    nonsilent_ranges = detect_nonsilent(
        audio,
        min_silence_len=200,    # im lặng >=200 ms
        silence_thresh=-40      # dưới -40 dBFS tính là im lặng
    )

    if not nonsilent_ranges:
        raise ValueError("Không tìm thấy đoạn âm thanh nào!")

    # Lấy vị trí bắt đầu của đoạn âm thanh đầu tiên
    start_ms = nonsilent_ranges[0][0]

    # Tính end_ms cho 5 giây sau đó
    end_ms = start_ms + 5_000

    # 3. Cắt 5 giây từ start_ms
    clip = audio[start_ms:end_ms]

    # 4. Xuất file kết quả
    clip.export(f"./sound-fixed/{type}/F{count}-{name[:-4]}.wav", format="wav") # bỏ đuôi .mp3 
    clip.export(f"./sound-fixed/{type}/FS{count}-{name}", format="mp3")
```

```python
def fix_all_mp3 (types):
    count = 1
    for type in types:
        print (f"Đang cắt files cho folder {type}") 
        for name in types[type]:
            print (f"file:{count}") 
            cut_mp3 (type, name, count)
            count +=1
        print (f"Đã cắt xong cho folder {type}") 

            
```
